```python
square = lambda x: x * x

try:
    numbers = range(1, 21)

    squares = [square(num) for num in numbers]

    even_squares = [num for num in squares if num % 2 == 0]

    print("Squares:", squares)
    print("Even Squares:", even_squares)

except Exception as e:
    print("An error occurred:", e)
```
