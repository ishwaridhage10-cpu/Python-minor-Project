```python
# Function to manage student marks
def manage_marks():
    marks = []  # Empty list to store marks

    # Taking 5 subject marks from the user
    try:
        for i in range(5):
            mark = float(input(f"Enter marks for Subject {i + 1}: "))
            marks.append(mark)

        # Calculating average marks
        average = sum(marks) / len(marks)

        # Finding highest and lowest marks
        highest = max(marks)
        lowest = min(marks)

        # Sorting marks in descending order
        marks.sort(reverse=True)

        # Displaying results
        print("\n--- Result ---")
        print("Average Marks:", average)
        print("Highest Marks:", highest)
        print("Lowest Marks:", lowest)
        print("Marks in Descending Order:", marks)

    # Handling invalid input
    except ValueError:
        print("Error: Please enter only numeric values.")


# Calling the function
manage_marks()
```
